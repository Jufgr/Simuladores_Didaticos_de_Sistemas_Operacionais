Simulador simples dos algoritmos FCFS, SSTF e SCAN.

```
# Escalonamento de Disco
# Algoritmos: FCFS, SSTF e SCAN
# Código feito para rodar no Jupyter Notebook ou no terminal Python.

def calcular_movimento(ordem, inicio):
    total = 0
    posicao_atual = inicio

    for cilindro in ordem:
        total += abs(cilindro - posicao_atual)
        posicao_atual = cilindro

    return total


def fcfs(inicio, requisicoes):
    ordem = requisicoes[:]
    total = calcular_movimento(ordem, inicio)
    return ordem, total


def sstf(inicio, requisicoes):
    pendentes = requisicoes[:]
    ordem = []
    posicao_atual = inicio

    while len(pendentes) > 0:
        mais_proximo = pendentes[0]

        for cilindro in pendentes:
            if abs(cilindro - posicao_atual) < abs(mais_proximo - posicao_atual):
                mais_proximo = cilindro

        ordem.append(mais_proximo)
        pendentes.remove(mais_proximo)
        posicao_atual = mais_proximo

    total = calcular_movimento(ordem, inicio)
    return ordem, total


def scan(inicio, requisicoes, tamanho_disco, direcao):
    menores = []
    maiores = []

    for cilindro in requisicoes:
        if cilindro < inicio:
            menores.append(cilindro)
        else:
            maiores.append(cilindro)

    menores.sort(reverse=True)
    maiores.sort()

    if direcao == "direita":
        ordem = maiores + [tamanho_disco - 1] + menores
    else:
        ordem = menores + [0] + maiores

    total = calcular_movimento(ordem, inicio)
    return ordem, total


def mostrar_resultado(nome, ordem, total):
    print("\n" + nome)
    print("Ordem de atendimento:", ordem)
    print("Total de cilindros percorridos:", total)


# -----------------------------
# Dados de teste
# -----------------------------

posicao_inicial = 53
fila_requisicoes = [98, 183, 37, 122, 14, 124, 65, 67]
tamanho_disco = 200
direcao_scan = "direita"

print("ESCALONAMENTO DE DISCO")
print("Posição inicial do braço:", posicao_inicial)
print("Fila de requisições:", fila_requisicoes)

ordem_fcfs, total_fcfs = fcfs(posicao_inicial, fila_requisicoes)
ordem_sstf, total_sstf = sstf(posicao_inicial, fila_requisicoes)
ordem_scan, total_scan = scan(posicao_inicial, fila_requisicoes, tamanho_disco, direcao_scan)

mostrar_resultado("FCFS", ordem_fcfs, total_fcfs)
mostrar_resultado("SSTF", ordem_sstf, total_sstf)
mostrar_resultado("SCAN", ordem_scan, total_scan)

resultados = {
    "FCFS": total_fcfs,
    "SSTF": total_sstf,
    "SCAN": total_scan
}

melhor = min(resultados, key=resultados.get)

print("\nComparação final")
print("Algoritmo mais eficiente neste teste:", melhor)
print("Menor deslocamento:", resultados[melhor], "cilindros")
```
