```python
# ALGORITMO DO BANQUEIRO - ESTADO SEGURO


def calcular_necessidade(maximo, alocado):
    """
    Calcula a matriz Need:
    Need = Max - Allocation
    """
    necessidade = []

    for i in range(len(maximo)):
        linha = []

        for j in range(len(maximo[i])):
            linha.append(maximo[i][j] - alocado[i][j])

        necessidade.append(linha)

    return necessidade


def verificar_estado_seguro(alocado, necessidade, disponivel):
    """
    Verifica se o sistema está em estado seguro.
    Retorna:
        True e a sequência segura, caso exista.
        False e lista vazia, caso contrário.
    """

    num_processos = len(alocado)

    trabalho = disponivel.copy()
    finalizado = [False] * num_processos
    sequencia_segura = []

    while len(sequencia_segura) < num_processos:

        encontrou_processo = False

        for i in range(num_processos):

            if not finalizado[i]:

                pode_executar = True

                for j in range(len(disponivel)):

                    if necessidade[i][j] > trabalho[j]:
                        pode_executar = False
                        break

                if pode_executar:

                    
                    for j in range(len(disponivel)):
                        trabalho[j] += alocado[i][j]

                    finalizado[i] = True
                    sequencia_segura.append(f"P{i}")
                    encontrou_processo = True

        if not encontrou_processo:
            return False, []

    return True, sequencia_segura


def imprimir_matriz(nome, matriz):

    print(f"\n{nome}")

    for i, linha in enumerate(matriz):
        print(f"P{i}: {linha}")


def main():

    alocado = [
        [0, 1, 0],
        [2, 0, 0],
        [3, 0, 2],
        [2, 1, 1],
        [0, 0, 2]
    ]

# Matriz Max
    maximo = [
        [7, 5, 3],
        [3, 2, 2],
        [9, 0, 2],
        [2, 2, 2],
        [4, 3, 3]
    ]
    disponivel = [3, 3, 2]

    necessidade = calcular_necessidade(maximo, alocado)

    print("=" * 50)
    print("ALGORITMO DO BANQUEIRO")
    print("=" * 50)

    imprimir_matriz("MATRIZ DE ALOCAÇÃO", alocado)
    imprimir_matriz("MATRIZ MÁXIMA", maximo)
    imprimir_matriz("MATRIZ DE NECESSIDADE", necessidade)

    print("\nRECURSOS DISPONÍVEIS")
    print(disponivel)

    seguro, sequencia = verificar_estado_seguro(
        alocado,
        necessidade,
        disponivel
    )

    print("\nRESULTADO")

    if seguro:
        print("O sistema está em ESTADO SEGURO.")
        print("Sequência segura encontrada:")
        print(" -> ".join(sequencia))
    else:
        print("O sistema está em ESTADO INSEGURO.")


if __name__ == "__main__":
    main()
```
